```python
import speech_recognition as aa
import pyttsx3
import pywhatkit
import datetime
import wikipedia
import googlesearch
import os
import pyautogui
import requests
from googlesearch import search
import schedule
import time
import calendar

listener = aa.Recognizer()

machine = pyttsx3.init()

def talk(text):
    machine.say(text)
    machine.runAndWait()

def input_instruction():
    global instruction
    try:
        with aa.Microphone() as origin:
            print("listening")
            speech = listener.listen(origin)
            instruction = listener.recognize_google(speech)
            instruction = instruction.lower()
            if "JAY" in instruction:
                instruction = instruction.replace('JAY'," ")
                print(instruction)
            print(instruction)
    except:
        pass
    return instruction

def remind_task(task, time):
    schedule.every().day.at(time).do(talk, f"Reminder: {task}")  

def input_reminder():
    instruction = input_instruction()
    if "remind me" in instruction:
        task = instruction.replace('remind me', "")
        time = input("Enter the time for the reminder (HH:MM): ")
        remind_task(task, time)
        talk("Reminder set for " + time)

def add_calendar_task():
    instruction = input_instruction()
    if "add task" in instruction:
        task = instruction.replace('add task', "")
        date = input("Enter the date for the task (DD/MM/YYYY): ")
        time = input("Enter the time for the task (HH:MM): ")
        year, month, day = map(int, date.split('/'))
        schedule.every().year.at(time).do(talk, f"Task: {task} on {calendar.month_name[month]} {day}, {year}")  
        talk("Task added to calendar for " + date + " at " + time)

def open_calendar():
    talk("Opening calendar")
    if os.name == 'nt':
        os.system('start outlookcal:')
    elif os.name == 'posix':
        os.system('open ical:')
    else:
        talk("Sorry, I don't know how to open the calendar on this system")

def play_JAY():
    while True:
        instruction = input_instruction()
        print(instruction)
        if "play" in instruction:
            song = instruction.replace('play'," ")
            talk("playing" + song)
            pywhatkit.playonyt(song)

        elif 'time' in instruction:
            time = datetime.datetime.now().strftime('%I:%M%p')
            talk('current time' + time)

        elif 'date' in instruction:
            date = datetime.datetime.now().strftime('%d /%m /%Y')
            talk("Today's date" + date)

        elif 'Yes sire how I can help you' in instruction:
            talk('Defenitely I will show your request')

        elif 'What is your name' in instruction:
            talk('I am JAY, What can I do for you?')

        elif 'who' in instruction:
            human = instruction.replace('who', "")
            info = wikipedia.summary(human,1)
            print(info)
            talk(info)
            
        elif 'search' in instruction:
            query = instruction.replace('search', "")
            talk("Searching for " + query)
            pywhatkit.search(query)
            talk("Search results are open")

        elif 'settings' in instruction:
            talk("Opening system settings")
            os.system('start ms-settings:')
            talk("System settings are open")

        elif 'change brightness' in instruction:
            talk("Changing brightness")
            pyautogui.press('winleft')
            pyautogui.typewrite('brightness')
            pyautogui.press('enter')
            talk("Brightness settings are open")

        elif 'change volume' in instruction:
            talk("Changing volume")
            pyautogui.press('winleft')
            pyautogui.typewrite('volume')
            pyautogui.press('enter')
            talk("Volume settings are open")

        elif 'change display settings' in instruction:
            talk("Changing display settings")
            pyautogui.press('winleft')
            pyautogui.typewrite('display settings')
            pyautogui.press('enter')
            talk("Display settings are open")

        elif 'calculate' in instruction:
            talk("Calculating")
            calculation = instruction.replace('calculate', " ")
            try:
                result = eval(calculation)
                talk("The result is " + str(result))
            except Exception as e:
                talk('JAY will help in')
                pywhatkit.search(instruction)
                talk("Showing the result on Google search")

        elif 'remind me' in instruction:
            input_reminder()

        elif 'add task' in instruction:
            add_calendar_task()

        elif 'open calendar' in instruction:
            open_calendar()

        else:
            talk('JAY will help in')
            pywhatkit.search(instruction)
            talk("Showing the result on Google search")

        schedule.run_pending() 
        time.sleep()  

play_JAY()
```
